# Classificar se aceita ou não o estágio, indicando sim ou não: (No Canvas)

```python
import tensorflow as tf

from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#@title Entrada de dados
X1 = np.array(
    [
        ['alto',   'longe', 'interessante'],
        ['baixo',  'perto', 'desinteressante'],
        ['baixo',  'longe', 'interessante'],
        ['alto',   'longe', 'desinteressante'],
        ['alto',   'perto', 'interessante'],
        ['baixo',  'longe', 'desinteressante']
    ]
)
Y = np.array( [ 'sim', 'não', 'sim', 'não', 'sim', 'não' ] )

X = X1
```

```python
#@title Altera os dados para números
for i in range(len(X1)):
  if X1[i][0] == "alto":
    X[i][0] = 0
  if X1[i][0] == "baixo":
    X[i][0] = 1
  if X1[i][1] == "longe":
    X[i][1] = 0
  if X1[i][1] == "perto":
    X[i][1] = 1
  if X1[i][2] == "interessante":
    X[i][2] = 0
  if X1[i][2] == "desinteressante":
    X[i][2] = 1


XX = pd.DataFrame(X, dtype=str, columns=['Salário', 'Localização', 'Função'])

YY = pd.DataFrame(Y, dtype=str, columns=['Decisão'])

XX


YY
```

```python
#@title Criando o classificador
arvore = DecisionTreeClassifier(criterion='entropy')
```

```python
#@title Criando a árvore
arvore = arvore.fit(XX, YY)

arvore.feature_importances_

arvore.classes_

arvore.score(XX, YY)

```

```python
#@title Visualizando essa árvore
from sklearn import tree
fig, ax = plt.subplots(figsize=(10,8))

tree.plot_tree(arvore)

plt.show()
```

```python
#@title Visualizando essa árvore 2
previsores = ['Salário', 'Localização', 'Função']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(arvore, feature_names=previsores, class_names = arvore.classes_, filled=True)
```

```python
#@title c) Classificar Salário=baixo, Localização=perto, Função=interessante
X3 = pd.DataFrame([['0','1','1']], columns=['Salário', 'Localização', 'Função'])
X3

previsoes3 = arvore.predict(X3)
previsoes3
```

```python
#@title d) Classificar Salário=alto, Localização=perto, Função=desinteressante
X2 = pd.DataFrame([['1','1','0']], columns=['Salário', 'Localização', 'Função'])
X2

previsoes = arvore.predict(X2)
previsoes
```

# Classificar se vai ou não jogar ténis, indicando sim ou não: (No Canvas)

```python
#@title Entrada de dados
X1 = np.array(
[["sol","quente","elevada","fraco"],
["sol","quente","elevada","forte"],
["nuvens","quente","elevada","fraco"],
["chuva","ameno","elevada","fraco"],
["chuva","fresco","normal","fraco"],
["chuva","fresco","normal","forte"],
["nuvens","fresco","normal","fraco"],
["sol","ameno","elevada","fraco"],
["sol","fresco","normal","fraco"],
["chuva","ameno","normal","forte"],
["sol","ameno","normal","forte"],
["nuvens","ameno","elevada","forte"],
["nuvens","quente","normal","fraco"],
["chuva","ameno","elevada","forte"]]
)

Y = np.array([
  'não',
  'não',
  'sim',
  'sim',
  'sim',
  'não',
  'sim',
  'não',
  'sim',
  'sim',
  'sim',
  'sim',
  'sim',
  'não',
])

X = X1

```

```python
#@title Altera os dados para números
for i in range(len(X1)):
  if X1[i][0] == "sol":
    X[i][0] = 0
  if X1[i][0] == "nuvens":
    X[i][0] = 1
  if X1[i][0] == "chuva":
    X[i][0] = 2
  if X1[i][1] == "quente":
    X[i][1] = 0
  if X1[i][1] == "ameno":
    X[i][1] = 1
  if X1[i][1] == "fresco":
    X[i][1] = 3
  if X1[i][2] == "elevada":
    X[i][2] = 0
  if X1[i][2] == "normal":
    X[i][2] = 1
  if X1[i][3] == "fraco":
    X[i][3] = 0
  if X1[i][3] == "forte":
    X[i][3] = 1

XX = pd.DataFrame(X, dtype=str, columns=['aspecto', 'temp', 'humidade', 'vento'])

YY = pd.DataFrame(Y, dtype=str, columns=['jogar tenis'])

XX


YY
```

```python
#@title Criando o classificador
arvore = DecisionTreeClassifier(criterion='entropy')
```

```python
#@title Criando a árvore
arvore = arvore.fit(XX, YY)

arvore.feature_importances_

arvore.classes_

arvore.score(XX, YY)

```

```python
#@title Visualizando essa árvore
from sklearn import tree
fig, ax = plt.subplots(figsize=(10,8))

tree.plot_tree(arvore)

plt.show()
```

```python
#@title Visualizando essa árvore 2
previsores = ['aspecto', 'temp', 'humidade', 'vento']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(arvore, feature_names=previsores, class_names = arvore.classes_, filled=True)
```

```python
#@title c) Classificar Aspecto=Sol, Temp.=Ameno, Humidade=Normal, Vento=Forte
X3 = pd.DataFrame([[0, 1, 1, 1]], columns=['aspecto', 'temp', 'humidade', 'vento'])
X3

previsoes3 = arvore.predict(X3)
previsoes3
```

```python
#@title d) Classificar Aspecto=Chuva, Temp.=Quente, Humidade=Normal, Vento=Fraco
X4 = pd.DataFrame([[2, 0, 1, 0]], columns=['aspecto', 'temp', 'humidade', 'vento'])
X4

previsoes4 = arvore.predict(X4)
previsoes4
```

```python
#@title Entrada de dados
X = np.array(
 [
 ['sol', 'alta', 'alta', 'não'],
 ['sol', 'alta', 'alta', 'sim'],
 ['nublado', 'alta', 'alta', 'não'],
 ['chuva', 'alta', 'alta', 'não'],
 ['chuva', 'baixa', 'normal', 'não'],
 ['chuva', 'baixa', 'normal', 'sim'],
 ['nublado', 'baixa', 'normal', 'sim'],
 ['sol', 'suave', 'alta', 'não'],
 ['sol', 'baixa', 'normal', 'não'],
 ['chuva', 'suave', 'normal', 'não'],
 ['sol', 'suave', 'normal', 'sim'],
 ['nublado', 'suave', 'alta', 'sim'],
 ['nublado', 'alta', 'normal', 'não'],
 ['chuva', 'suave', 'alta', 'sim']
 ]
)
Y = np.array( [ 'não joga', 'não joga', 'joga', 'joga', 'joga', 'não joga', 'joga',
'não joga', 'joga', 'joga', 'joga', 'joga', 'joga', 'não joga' ] )
```

```python
#@title Tratando dados
X1 = pd.DataFrame(X, dtype=str, columns=['Céu', 'Temperatura',
'Umidade', 'Vento'])
XX = pd.get_dummies(data=X1, columns=['Céu', 'Temperatura',
'Umidade', 'Vento'])
XX
Y1 = pd.DataFrame(Y, dtype=str, columns=['Classe'])
YY = pd.get_dummies(data=Y1, columns=['Classe'])
YY
```

```python
#@title Criando o classificador
arvore = DecisionTreeClassifier(criterion='entropy', max_depth=2)
```

```python
#@title Criando a árvore
arvore = arvore.fit(XX, YY)
arvore.feature_importances_
arvore.classes_
arvore.score(XX, YY)
```

```python
#@title Visualizando essa árvore
from sklearn import tree
fig, ax = plt.subplots(figsize=(10,8))
tree.plot_tree(arvore)
plt.show()
```

```python
#@title Visualizando essa árvore 2
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(arvore, feature_names=XX.columns, class_names = ['Não Joga', 'Joga'], filled=True)
```

```python
# Instala a biblioteca
!pip install numpy
!pip install matplotlib
!pip install tensorflow
!pip install scikeras
# Importa rotinas
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pydotplus
import matplotlib.image as mpimg
```

```python
#@title Load Data
data = pd.read_csv('/content/drug200.csv')
data.head()
```

```python
#@title Dependent and Independent variables
X = data[['Age','Sex','BP','Cholesterol','Na_to_K']].values
y = data[['Drug']].values # Dependent variable
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1])
le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW','NORMAL','HIGH'])
X[:,2] = le_BP.transform(X[:,2])
le_Chol = preprocessing.LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3])
```

```python
# Train and Test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
random_state=1)
```

```python
# Define Model
model = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
```

```python
# Fit Model
model.fit(X_train,y_train)
model.feature_importances_
model.classes_
model.score(X_train, y_train)
```

```python
# Prediction
predTree = model.predict(X_test)
```

```python
# Accuracy
print("Decision Trees's Accuracy: ", metrics.accuracy_score(y_test,
predTree))
```

```python
# Visualizando essa árvore
from sklearn import tree
fig, ax = plt.subplots(figsize=(10,8))
tree.plot_tree(model)
plt.show()
```

```python
# Visualizando essa árvore 2
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(model, feature_names=data.columns[0:5], class_names =
np.unique(y_train), filled=True)
```

```python
# Regra Se- Então
from sklearn.tree import export_text
r = export_text(model, feature_names=['Age', 'Sex', 'BP', 'Cholesterol',
'Na_to_K'])
print(r)
```

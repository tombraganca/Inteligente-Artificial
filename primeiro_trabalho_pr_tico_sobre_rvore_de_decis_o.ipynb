```python
import tensorflow as tf

from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
entradas = np.array(
 [
 ['alto', 'longe', 'interessante'],
 ['baixo', 'perto', 'desinteressante'],
 ['baixo', 'longe', 'interessante'],
 ['alto', 'longe', 'desinteressante'],
 ['alto', 'perto', 'interessante'],
 ['baixo', 'longe', 'desinteressante']
 ]
)

saidas = np.array( [ 'sim', 'não', 'sim', 'não', 'sim', 'não' ] )
```

```python
#@title Ajusta as informações para numérico
entrada_ajustada = entradas
for i in range(len(entradas)):
 if entradas[i][0] == "alto":
  entrada_ajustada[i][0] = 0
 if entradas[i][0] == "baixo":
  entrada_ajustada[i][0] = 1
 if entradas[i][1] == "longe":
  entrada_ajustada[i][1] = 0
 if entradas[i][1] == "perto":
  entrada_ajustada[i][1] = 1
 if entradas[i][2] == "interessante":
  entrada_ajustada[i][2] = 0
 if entradas[i][2] == "desinteressante":
  entrada_ajustada[i][2] = 1

tabela_de_entrada = pd.DataFrame(entrada_ajustada, dtype=str, columns=['Salário','Localização', 'Função'])
tabela_de_saida = pd.DataFrame(saidas, dtype=str, columns=['Decisão'])
```

```python
#@title Criando o classificador e a arvore
arvore = DecisionTreeClassifier(criterion='entropy')
arvore = arvore.fit(tabela_de_entrada, tabela_de_saida)
```

```python
#@title Visualizando essa árvore
from sklearn import tree
fig, ax = plt.subplots(figsize=(10,8))
tree.plot_tree(arvore)
plt.show()
```

```python
#@title a) Classificar Salário=baixo, Localização=perto, Função=interessante
X2 = pd.DataFrame([['0','1','1']], columns=['Salário', 'Localização', 'Função'])
X2
```

```python
#@title Resposta
previsoes = arvore.predict(X2)
previsoes
```

```python
#@title d) Classificar Salário=alto, Localização=perto, Função=desinteressante
X3 = pd.DataFrame([['1','1','0']], columns=['Salário','Localização', 'Função'])
X3
```

```python
#@title Resposta
previsoes3 = arvore.predict(X3)
previsoes3
```
